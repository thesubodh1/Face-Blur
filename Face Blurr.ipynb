```python
import cv2
import os
import mediapipe as mp

# read image
img = cv2.imread("subodh1.jpg")



H,W,_ = img.shape

# detect face
mp_face_detection = mp.solutions.face_detection

# model selection = (0 in 1) (0 = close to camera ,1 = 5m far from camera)
#min_detection = 50% 

with mp_face_detection.FaceDetection(model_selection=0,min_detection_confidence=0.5) as face_detection: 
    img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    out  = face_detection.process(img_rgb)

    print(out.detections)

    if out.detections is not None:
        for detection in out.detections:
            location_data = detection.location_data
            bbox = location_data.relative_bounding_box
    
            x1,y1,w,h = bbox.xmin, bbox.ymin, bbox.width, bbox.height

            x1 = int(x1*W)
            y1 = int(y1*H)
            w = int(w*W)
            h = int(h*H)

            # img_reg = cv2.rectangle(img,(x1,y1),(x1+w,y1+h),(0,255,0),10)

            # blur faces
            img[y1:y1+h,x1:x1+w,:] = cv2.blur(img[y1:y1+h,x1:x1+w,:],(30,20))

            
        
        


cv2.imshow("frame",img)
cv2.imwrite("blurred_img.jpg",img)
cv2.waitKey(0)
```
